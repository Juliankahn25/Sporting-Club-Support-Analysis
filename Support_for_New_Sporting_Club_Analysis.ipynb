```python
# Step 1: Install dependencies (if needed)
!pip install pandas scipy --quiet

# Step 2: Import required libraries
import pandas as pd
from scipy.stats import spearmanr

# Step 3: Upload your CSV file
from google.colab import files
uploaded = files.upload()

# Step 4: Load the file (replace the filename if needed)
df = pd.read_csv("2025 Sporting JAX - Final Data.csv", encoding="ISO-8859-1")

# Step 5: Extract relevant columns and handle invalid codes (98 = Don’t know, 99 = Refusal)
fan = df['FAN'].apply(lambda x: x if x not in [98, 99] else None)
important = df['IMPORTANT'].apply(lambda x: x if x not in [98, 99] else None)

# Step 6: Drop missing values
valid_df = pd.DataFrame({'FAN': fan, 'IMPORTANT': important}).dropna()

# Step 7: Compute Spearman correlation
corr, p_value = spearmanr(valid_df['FAN'], valid_df['IMPORTANT'])

# Step 8: Print results
print("Spearman Correlation:", corr)
print("P-Value:", p_value)
```

Correlation ranges from -1 to +1:

Closer to -1 = negative relationship

Closer to +1 = positive relationship

Around 0 = no relationship

P-value < 0.05 = statistically significant correlation.

This segment of the project uses Spearman correlation analysis to examine the potential relationship between two key survey responses: how strongly individuals identify as sports fans (FAN) and how important they believe professional women's sports are (IMPORTANT). By quantifying the association between these two ordinal variables, the analysis seeks to determine whether a meaningful, monotonic relationship exists—essentially asking: do stronger sports fans also place higher importance on women's professional sports?

To ensure data quality, the code first filters out invalid responses coded as “Don’t know” or “Refusal” (values 98 and 99), replacing them with None so they can be excluded from the analysis. After cleaning the data, a Spearman correlation is computed using Python’s scipy.stats library. Spearman’s method is appropriate here because both variables are ranked rather than continuous. The resulting correlation coefficient (ranging from -1 to 1) reveals the strength and direction of the association, while the p-value indicates whether the relationship is statistically significant.

This analysis provides deeper insight into public alignment between general sports enthusiasm and attitudes toward women's sports. A strong positive correlation would suggest that efforts to increase general fan engagement could naturally support the growth of women’s sports—information that is particularly valuable for marketing teams, league organizers, and community outreach initiatives. Additionally, this portion of the project serves as an example of how Python can be used to carry out inferential statistical methods, strengthening the analytical rigor of sports-based research.

```python
# Install necessary packages if needed
!pip install pandas seaborn matplotlib scipy --quiet

# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
from google.colab import files

# Upload CSV
uploaded = files.upload()

# Load data
df = pd.read_csv("2025 Sporting JAX - Final Data.csv", encoding="ISO-8859-1")

# Clean and filter FAN and IMPORTANT columns
df_clean = df[['FAN', 'IMPORTANT']].copy()
df_clean = df_clean[(~df_clean['FAN'].isin([98, 99])) & (~df_clean['IMPORTANT'].isin([98, 99]))].dropna()

# Convert to integers
df_clean['FAN'] = df_clean['FAN'].astype(int)
df_clean['IMPORTANT'] = df_clean['IMPORTANT'].astype(int)

# Calculate Spearman correlation
corr, p_val = spearmanr(df_clean['FAN'], df_clean['IMPORTANT'])

# Set plot style
sns.set(style="whitegrid")

# Create a violin plot with boxplot overlay
plt.figure(figsize=(10, 6))
sns.violinplot(x='FAN', y='IMPORTANT', data=df_clean, inner=None, palette="Pastel1")
sns.boxplot(x='FAN', y='IMPORTANT', data=df_clean, width=0.2, boxprops=dict(alpha=0.5), showcaps=True, showfliers=False)

# Titles and labels
plt.title(f'Relationship Between Sports Fandom and Support for Women’s Soccer Team\nSpearman r = {corr:.2f}, p = {p_val:.3f}')
plt.xlabel("FAN (1=Strongly agree → 4=Strongly disagree)")
plt.ylabel("IMPORTANT (1=Very important → 4=Not at all important)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Violin plot: Shows full distribution density for IMPORTANT across FAN levels. (Tops show small percetanges of answers either being 'Don't Know' or 'Refusal' (88 or 99))

This analysis explores the relationship between how strongly individuals identify as sports fans and how important they believe a professional women’s sports team is to the local community. The data is drawn from a larger survey and specifically focuses on two key ordinal variables: FAN (which measures agreement with being a sports fan) and IMPORTANT (which measures the perceived importance of the team). Before analysis, invalid responses (e.g., “Don’t know” and “Refusal”) are removed to ensure that only meaningful data contributes to the results. The values are then cleaned and converted to integers for proper statistical handling.

To assess the strength of the relationship between these two variables, the code computes a Spearman correlation coefficient, which is ideal for ranked or ordinal data. This provides a measure of the monotonic association between fan identity and team importance—meaning it reveals whether higher fan enthusiasm generally aligns with stronger support for the women’s team. A statistically significant Spearman r and a low p-value would indicate that those who consider themselves bigger sports fans also tend to value the presence of the team more highly.

To complement the statistical result, the project includes a violin plot overlaid with a boxplot. This hybrid visualization not only displays the distribution of responses for each level of fandom but also highlights medians, interquartile ranges, and density curves—making it easy to visually interpret trends. Together, the statistical output and the plot offer a compelling narrative about how community sentiment toward women’s professional sports may be shaped by overall sports enthusiasm. This analysis demonstrates not only the power of data visualization and correlation but also practical use of Python and Seaborn in applied social research.

```python
# Step 1: Install required libraries
!pip install pandas matplotlib seaborn --quiet

# Step 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Step 3: Upload the CSV file
uploaded = files.upload()

# Step 4: Load your survey data (adjust file name if needed)
df = pd.read_csv("2025 Sporting JAX - Final Data.csv", encoding="ISO-8859-1")

# Step 5: Define INFLUENCE column-to-label mapping
influence_labels = {
    'INFLUENCE_1': "Ticket prices",
    'INFLUENCE_2': "Game time/Day of week",
    'INFLUENCE_3': "Location of stadium",
    'INFLUENCE_4': "Quality of competition",
    'INFLUENCE_5': "Game promotions/giveaways",
    'INFLUENCE_6': "Food and drink options",
    'INFLUENCE_7': "Family-friendliness",
    'INFLUENCE_8': "Public transportation access",
    'INFLUENCE_9': "Parking and accessibility",
    'INFLUENCE_10': "Star players",
    'INFLUENCE_11': "Weather",
    'INFLUENCE_12': "Other"
}

# Step 6: Subset just the INFLUENCE columns
influence_cols = list(influence_labels.keys())
influence_df = df[influence_cols]

# Step 7: Count how many times each factor was selected
influence_counts = (influence_df == 1).sum().rename(index=influence_labels).sort_values(ascending=False)

# Step 8: Create a beautiful horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=influence_counts.values, y=influence_counts.index, palette="magma")
plt.title("Most Influential Factors for Attending Sporting JAX Matches", fontsize=14, weight='bold')
plt.xlabel("Number of Respondents Who Selected", fontsize=12)
plt.ylabel("Influence Factor", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

This portion of the project aims to uncover the key factors that influence community members’ decisions to attend matches for the newly launched Sporting JAX soccer team. The survey collected responses across 12 potential influence categories—ranging from ticket prices and stadium location to public transportation access and star players. Participants were allowed to select multiple factors they considered important. This analysis identifies which of those factors were selected most frequently, providing a clearer understanding of what drives local game attendance.

The code begins by mapping the survey’s coded column names (e.g., INFLUENCE_1, INFLUENCE_2) to their real-world labels (like "Ticket prices" or "Weather"). It then calculates the number of times each factor was marked as influential (value = 1) across all responses. By summing these values for each category, the analysis ranks which elements of the game-day experience are most important to potential attendees. This structured approach allows for efficient handling of multi-response survey data and creates actionable insights for event organizers and team marketing staff.

The final output is a horizontal bar chart that visualizes the top attendance-driving factors. Presented with a bold, color-graded design using Seaborn, the chart helps stakeholders quickly grasp which areas matter most to their audience. For example, if “Ticket prices,” “Family-friendliness,” and “Game promotions” are among the most selected, then those aspects should become key focus areas for outreach, pricing strategy, and game-day planning. This analysis is a strong example of how Python can be used to translate survey feedback into strategic decisions that improve fan engagement and boost attendance.

```python
# Step 1: Install dependencies
!pip install pandas matplotlib seaborn --quiet

# Step 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Step 3: Upload CSV file
uploaded = files.upload()

# Step 4: Load dataset
df = pd.read_csv("2025 Sporting JAX - Final Data.csv", encoding="ISO-8859-1")

# Step 5: Map codes to labels
attend_map = {
    1: "Very likely", 2: "Somewhat likely", 3: "Somewhat unlikely",
    4: "Very unlikely", 98: "Don't know", 99: "Refusal"
}
streamsvc_map = {1: "Yes", 2: "No", 98: "Don't know", 99: "Refusal"}
streams_map = {
    1: "Very likely", 2: "Somewhat likely", 3: "Somewhat unlikely",
    4: "Very unlikely", 98: "Don't know", 99: "Refusal"
}

# Step 6: Apply label mappings
df['ATTEND_Label'] = df['ATTEND'].map(attend_map)
df['STREAMSVC_Label'] = df['STREAMSVC'].map(streamsvc_map)
df['STREAMSJC_Label'] = df['STREAMSJC'].map(streams_map)

# Step 7: Count responses
attendance_counts = df['ATTEND_Label'].value_counts().sort_index()
streamsvc_counts = df['STREAMSVC_Label'].value_counts().sort_index()
streams_counts = df['STREAMSJC_Label'].value_counts().sort_index()

# Step 8: Combine into a DataFrame
engagement_df = pd.DataFrame({
    'Attend Likelihood': attendance_counts,
    'Subscribed to Sports Streaming Service': streamsvc_counts,
    'Likelihood to Stream Sporting JAX': streams_counts
}).fillna(0).astype(int)

# Step 9: Transpose and plot with larger formatting
engagement_df_t = engagement_df.T

plt.figure(figsize=(16, 8))
bar_plot = engagement_df_t.plot(kind='bar', figsize=(16, 8), colormap='viridis', width=0.8)

plt.title("Attendance vs. Streaming Preferences", fontsize=18, weight='bold')
plt.ylabel("Number of Respondents", fontsize=14)
plt.xlabel("Engagement Type", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.legend(title="Response Category", fontsize=12, title_fontsize=13)
plt.tight_layout()
plt.show()
```

This portion of the analysis evaluates public engagement with Sporting JAX by examining three interconnected areas: how likely participants are to attend games, whether they currently subscribe to sports streaming services, and their willingness to stream Sporting JAX matches in the future. The survey responses for these questions were originally coded numerically, so the first step is to convert those codes into meaningful labels such as "Very likely," "No," or "Don't know." This labeling improves the interpretability of the data and prepares it for visual analysis.

Once the labels are mapped, the code counts how many participants selected each response category for all three engagement questions. These counts are compiled into a single DataFrame and then transposed to make the data presentation more visually intuitive. The final output is a multi-category grouped bar chart, which clearly shows the distribution of responses across all three dimensions. This visualization enables quick comparisons between how people feel about attending games in person versus streaming them, as well as whether they already engage with sports content through digital platforms.

The insights from this chart are valuable for understanding consumer behavior and audience segmentation. For example, a large number of respondents who say they're "Very likely" to stream Sporting JAX games but aren't current streaming service subscribers could point to an opportunity for the club to launch its own digital platform or partner with an existing one. Similarly, high willingness to attend games in person might guide marketing and ticketing strategies.
